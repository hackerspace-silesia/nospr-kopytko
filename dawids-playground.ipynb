```python
%matplotlib inline 
from skimage.color import rgb2hsv
from skimage.io import imread, imshow, show

import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def show_images(data, small=False):
    fig = plt.figure(figsize=(20, 20))
    ldata = len(data)

    for i, (im, cmap) in enumerate(data, start=1):
        if small:
            ax = fig.add_subplot(1, ldata, i, xticks=[], yticks=[])
            ax.imshow(im, cmap=cmap)
        else:
            fig = plt.figure(figsize=(20, 20))
            imshow(im, cmap=cmap)
            show()
    if small:
        fig.show()
```

```python
img = imread("a.jpg")
plt.figure(figsize=(20, 20))
imshow(img)
show()
```

```python
negative = cv2.bitwise_not(img)
plt.figure(figsize=(20, 20))
imshow(negative)
```

```python
normalizedImg = np.zeros((800, 800))
normalized_image = cv2.normalize(negative,  normalizedImg, 0, 255, cv2.NORM_MINMAX)
plt.figure(figsize=(20, 20))
imshow(normalized_image)
```

```python
blurred_image = cv2.medianBlur(normalized_image, 25)
plt.figure(figsize=(20, 20))
imshow(blurred_image)
```

```python
hsv = rgb2hsv(blurred_image)
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

show_images([
    (h, 'hsv'),
    (s, 'Greys'),
    (v, 'gray'),
])
```
